```python
import sqlite3
import pandas as pd

# Create connection
conn = sqlite3.connect("bookstore.db")
cursor = conn.cursor()

print("✓ Database connection established")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS books(
    book_id INTEGER PRIMARY KEY AUTOINCREMENT,
    title TEXT NOT NULL,
    author TEXT NOT NULL,
    price REAL NOT NULL,
    stock_quantity INTEGER DEFAULT 0
)
""")

print("✓ Books table created successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS customers(
    customer_id INTEGER PRIMARY KEY AUTOINCREMENT,
    name TEXT NOT NULL,
    email TEXT UNIQUE NOT NULL,
    city TEXT,
    join_date TEXT
)
""")

print("✓ Customers table created successfully")
```

```python
cursor.execute("""
CREATE TABLE IF NOT EXISTS orders(
    order_id INTEGER PRIMARY KEY AUTOINCREMENT,
    customer_id INTEGER,
    book_id INTEGER,
    quantity INTEGER NOT NULL,
    order_date TEXT NOT NULL,
    total_amount REAL,
    FOREIGN KEY(customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY(book_id) REFERENCES books(book_id)
)
""")

print("✓ Orders table created successfully")
```

```python
tables = ["books","customers","orders"]

for table in tables:
    print(f"\nSchema for {table} table:")
    schema = cursor.execute(f"PRAGMA table_info({table})").fetchall()
    for row in schema:
        print(row)
```

```python
cursor.execute("SELECT COUNT(*) FROM orders")
print("Total Orders:", cursor.fetchone()[0])
```

```python
cursor.execute("""
SELECT customers.name, COUNT(orders.order_id) AS total_orders
FROM orders
JOIN customers
ON customers.customer_id=orders.customer_id
GROUP BY customers.name
ORDER BY total_orders DESC
LIMIT 1
""")

print(cursor.fetchone())
```

```python
cursor.execute("SELECT SUM(total_amount) FROM orders")
print("Total Revenue:", cursor.fetchone()[0])
```

```python
books_df = pd.read_sql("SELECT * FROM books", conn)
customers_df = pd.read_sql("SELECT * FROM customers", conn)
orders_df = pd.read_sql("SELECT * FROM orders", conn)

print(books_df.head(3))
print(customers_df.head(3))
print(orders_df.head(3))
```

```python
report = orders_df.merge(customers_df,on="customer_id") \
                  .merge(books_df,on="book_id")

report = report[['order_id','name','city','title','quantity','total_amount']]

report.columns=['Order ID','Customer Name','City','Book Title','Quantity','Total Amount']

print(report)
```

```python
avg = orders_df["total_amount"].mean()
print("Average Order Value:",avg)
```

```python
orders_by_city = report.groupby("City")["Order ID"].count()
print(orders_by_city)
```

```python
popular = report.groupby("Book Title")["Quantity"].sum().sort_values(ascending=False)
print(popular.head(1))
```

```python
discounts_data = {
    'book_id':[1,2,3,4,5],
    'discount_percent':[10,15,5,20,12]
}

discount_df = pd.DataFrame(discounts_data)
```

```python
discount_df.to_sql("discounts",conn,if_exists="replace",index=False)

print("✓ Discounts table created")
```

```python
print(pd.read_sql("SELECT * FROM discounts",conn))
```

```python
query = """
SELECT
books.title,
books.price AS original_price,
discounts.discount_percent,
books.price*(1-discounts.discount_percent/100) AS discounted_price
FROM books
JOIN discounts
ON books.book_id=discounts.book_id
"""

print(pd.read_sql(query,conn))
```
